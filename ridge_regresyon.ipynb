RİDGE REGRESYON
overfitting durumlarını dengelemek için kullanılır.

SSE + λ × Σ(β²)
Burada:

SSE: Hata karelerinin toplamı (standart regresyon hatasıdır)
λ (lambda): Ceza parametresi (regularizasyon katsayısı)
β: Model katsayıları

Lambda değeri arttıkça, katsayılar küçülmeye zorlanır. λ = 0 ise normal regresyon olur.


Niçin Kullanılır?
1. Multicollinearity (Çoklu Doğrusallık) Problemini Çözer
Bağımsız değişkenler arasında yüksek korelasyon olduğunda, standart regresyon katsayılar tahminini güvenilmez hale getirir. Ridge regresyon bu sorunu hafifletir.
2. Aşırı Uyum (Overfitting) Önler
Ceza terimi, modelin eğitim verisine çok fazla uymasını engeller. Böylece test verisi üzerinde daha iyi genelleme yeteneği sağlar.
3. Daha Kararlı Tahminler
Daha istikrarlı ve daha az değişken katsayı tahminleri sağlar. Küçük veri değişikliklerinde sonuçlar çok fazla değişmez.
4. Tüm Değişkenleri Modelde Tutar
Lasso regresyondan (L1) farklı olarak, ridge regresyon (L2) değişken seçimi yapmaz. Tüm değişkenlerin katsayısını azaltır ancak tamamen sıfıra çıkarmaz. Bu, tüm değişkenlerin potansiyel olarak faydalı olduğu durumlarda iyi bir seçenektir.



Kat sayılar üzerinde regülasyon yapılır. kat sayılar küçülür ama 0 olmaz. öznitelik (features) azalmaz.


lambda azaldıkça grafikteki fonksiyon çizgisi x eksenine paralelleşir. yaklaşır


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
df=pd.read_csv('student_scores.csv')
```

```python
df.head(5)
```

```python
y=df['Scores']
x=df[['Hours']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,8))
plt.scatter(x,y)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x,y)
```

```python
alphalar=[1,2,10,100,200]
for i in alphalar:
  r=Ridge(alpha=i)
  modelr=r.fit(x_train,y_train)
  print(f"alpha={i} score={modelr.score(x_test,y_test)} katsayılar={modelr.coef_}")
```
