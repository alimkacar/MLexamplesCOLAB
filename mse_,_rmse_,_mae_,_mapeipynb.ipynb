Makine Öğrenmesinde Hata Metrikleri
-----------------------------------------------------------------------------------------
Bu dört metrik, regresyon modellerinin performansını ölçmek için kullanılan tahmin hata ölçütleridir. Modelin gerçek değerlerden ne kadar saptığını gösterirler.

MSE (Mean Squared Error - Ortalama Kare Hata)
Hataların karesinin ortalamasıdır. Her hatayı karesi alarak çıkartılır ve ortalaması hesaplanır.
Büyük hataları daha fazla cezalandırdığı için aykırı değerlere (outliers) karşı hassastır. Sıfıra ne kadar yakınsa o kadar iyidir.


RMSE (Root Mean Squared Error - Kök Ortalama Kare Hata)
MSE'nin kareköküdür. MSE'yi orijinal veri birimine geri dönüştürerek daha yorumlanabilir hale getirir.
Örneğin fiyat tahmini yapıyorsanız, RMSE de para birimiyle ifade edilir. Büyük hataları cezalandırması nedeniyle aykırı değerlere duyarlıdır.


MAE (Mean Absolute Error - Ortalama Mutlak Hata)
Hataların mutlak değerlerinin ortalamasıdır.
Her hatanın işaretinden bağımsız olarak, sayısal büyüklüğünü dikkate alır. RMSE'den daha basit ve aykırı değerlere karşı daha dayanıklıdır.


MAPE (Mean Absolute Percentage Error - Ortalama Mutlak Yüzde Hata)
Hataların yüzde olarak ifadesidir. Gerçek değere göre yüzde kaç hata yapıldığını gösterir.
Farklı ölçeklerdeki veriler arasında karşılaştırma yapmak için idealdir. Örneğin, satış tahmini veya enerji tüketimi gibi yüzde cinsinden düşünmek daha anlamlı olabilir.


Pratik Örnek: Ev fiyatı tahmin modeliniz için:

Gerçek: 300.000 TL
Tahmin: 280.000 TL
Hata: -20.000 TL

MSE ve RMSE bu hatayı ağırlaştırır, MAE basitçe 20.000 TL olarak sayar, MAPE ise %6.67 hata olarak gösterir.

Hangi metrik kullanmanız gerekirse, sorunun doğasına ve aykırı değerlere karşı ne kadar toleranslı olmak istediğinize bağlıdır.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import root_mean_squared_error
```

```python
df=pd.read_csv("medicalcostpersonal.csv")
```

```python
df.head(5)
```

```python
df = pd.get_dummies(df, columns=["smoker", "region", "sex"], drop_first=True, dtype=int)
```

```python
df.head(5)
```

```python
df.info()
```

```python
y=df["charges"]
x=df.drop("charges", axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=45)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[21,20,0,1,0,0,1,1]])
```

```python
dfh=pd.DataFrame()
```

```python
dfh["y"]=y
```

```python
dfh.head(5)
```

```python
y_tahmin=model.predict(x)
```

```python
dfh['tahmin']=y_tahmin
```

```python
dfh.head(10)
```

```python
dfh['hata']=y-y_tahmin
```

```python
dfh.head(5)
```

```python
dfh['MSE']=dfh['hata']**2
```

```python
dfh.head(5)
```

```python
dfh['RMSE']=np.sqrt(dfh['MSE'])
```

```python
dfh.head(6)
```

```python
dfh['MAE']=np.abs(dfh['hata'])
```

```python
dfh.head(5)
```

```python
dfh['MAPE']=np.abs((y-y_tahmin)/y)
```

```python
dfh.head(6)
```

```python
dfh.mean()
```

```python
mean_absolute_error(y,y_tahmin)
```

```python
mean_absolute_percentage_error(y,y_tahmin)
```

```python
mean_squared_error(y,y_tahmin)
```

```python
root_mean_squared_error(y,y_tahmin)
```
