```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.DataFrame()
```

```python
df["cümleler"]=["bakkaldan ekmek al","bakkaldan yumurta al","kasaptan et al","manavdan patates al","manavdan ıspanak al","manavdan muz al","kasaptan tavuk al"]
```

```python
cv=CountVectorizer(max_features=3)
a=cv.fit_transform(df["cümleler"])
x=a.toarray()
```
