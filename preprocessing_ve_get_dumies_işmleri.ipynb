```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('audi.csv')
```

```python
df.head(5)
```

```python
df = df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score', 'PPY']) #istenmeyen sütunları kaldırmak için drop metodu kullanılır, Preprocessing işlemi
```

```python
df.head(5)
```

transmission fuel type gibi sütunlara kategorisel veri deniyor bir sayısal değer içermedikleri için.
year miles gibi verilere ise nümerik veri denir çünkü sayısal bir kaşılığı vardır.


```python
df.info() #oluşan tablonun sütunları hakkında bilgi veriyor
```

```python
df['Engine']=df['Engine'].str.replace("L", "") #eninge sütununu nümer veriye dönüştürmek için sayısal verilerin yanındaki hartfleri str.replace metoduyla kaldırdık
```

```python
df.head(3)
```

```python
df["Engine"]=pd.to_numeric(df["Engine"])
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['Type','Transmission','Fuel'],drop_first=True)
```

```python
df.head(3)
```

```python
dummy_cols = ['Transmission_Manual', 'Fuel_Petrol']
df[dummy_cols] = df[dummy_cols].astype(int)
```

```python
df.head(3)
```

```python
y=df["Price(£)"]
x=df.drop("Price(£)",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
lm = LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
