```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression,Ridge
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('kalpanaliz.csv')
df.head(5)
```

```python
df.columns
```

```python
df = df.drop(['id','dataset'], axis=1)
```

```python
df.head(10)
```

```python
df=pd.get_dummies(df,columns=['sex','cp','fbs','restecg','exang','slope','thal'],drop_first=True)
```

```python
df = df.fillna(0).astype(int)
#.fillna(0) - Eksik deerleri doldur
#astype(int) - Tm stunlar integer yap
```

```python
df.columns
```

```python
y=df['num']
x=df.drop('num',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.77,random_state=45)
```

```python
lr=LogisticRegression()
model=lr.fit(x_train,y_train)
model.score(x_test,y_test)
```
